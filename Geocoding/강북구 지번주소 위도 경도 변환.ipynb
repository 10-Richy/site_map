```python
import pandas as pd
df = pd.read_csv('C:\py_temp\Richy\서울특별시_강북구_공사장 현황.csv',encoding='cp949')
df.head()

#df크기확인
df.shape

#주소데이터 전처리 함수
def get_addrs(x):
    x1 = x.split(' ')
    return " ".join(x1[2:4])
    
# 주소 전처리
df['지번주소'] = df['대지위치'].apply(get_addrs)
df
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

#위도 반환 함수
def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude

    except:
        return 0
        
#경도 반환 함수
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude

    except:
        return 0

# 변환
df['위도'] = df['지번주소'].apply(geocoding_lat)
df['경도'] = df['지번주소'].apply(geocoding_lon)
df
```

```python


#csv저장
df.to_csv('C:\py_temp\Richy\강북구민간geopy_address.csv',encoding='cp949',index=False)

```
