```python
import pandas as pd
df = pd.read_csv('C:\py_temp\Richy\서울특별시 영등포구_민간공사장현황_20221102.csv',encoding='cp949')
df.head()

#df크기확인
df.shape

#주소데이터 전처리 함수
def get_addrs(x):
    x1 = x.split(' ')
    return " ".join(x1[2:4])
    
# 주소 전처리
df['지번주소'] = df['대지위치'].apply(get_addrs)
df
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

#위도 반환 함수
def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude

    except:
        return 0
        
#경도 반환 함수
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude

    except:
        return 0

# 변환
df['위도'] = df['지번주소'].apply(geocoding_lat)
df['경도'] = df['지번주소'].apply(geocoding_lon)
df
```

```python
#위치 정보를 제대로 불러오지 못한 부분 지번주소로 불러와 대체하기
#0값을 가지는 행 번호 추출하여 리스트 생성
nulls = df[df['위도'] == 0]
null_index = list(nulls.index)
null_index

#해당 index 부분만 지번번호 사용하여 주소정보 재추출
for i in null_index:
    df['위도'][i] = geocoding_lat(df['지번주소'][i])
    df['경도'][i] = geocoding_lon(df['지번주소'][i])
    
df[df['위도'] == 0]

#csv저장
df.to_csv('C:\py_temp\Richy\금천구구민간geopy_address.csv',encoding='cp949',index=False)

```
